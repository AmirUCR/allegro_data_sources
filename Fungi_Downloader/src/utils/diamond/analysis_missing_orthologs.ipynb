```python
import os
import pandas as pd

acc_to_name = {
    'NP_009673.1': 'LYS2',
    'NP_010290.3': 'TRP1',
    'NP_010851.1': 'CAN1',
    'NP_010893.3': 'URA3',
    'NP_012965.3': 'GAP1',
    'NP_013406.1': 'MET17',
    'NP_015387.1': 'FCY1',
}

num_files = len(os.listdir('outputs'))

ortho = pd.read_csv("Orthogroups.tsv", sep='\t')
ortho        
```

```python
ortho['yarrowia_lipolytica']
```

```python
def find_cols_with_x_not_found(df, x):
    return df.columns[df.eq('Not found').sum() == x].tolist()

# Identify columns where only one row is "Not Found"
cols_with_one_not_found = find_cols_with_x_not_found(ortho, 1)
cols_with_two_not_found = find_cols_with_x_not_found(ortho, 2)
cols_with_three_not_found = find_cols_with_x_not_found(ortho, 3)
cols_with_four_not_found = find_cols_with_x_not_found(ortho, 4)
cols_with_five_not_found = find_cols_with_x_not_found(ortho, 5)
cols_with_six_not_found = find_cols_with_x_not_found(ortho, 6)
cols_with_seven_not_found = find_cols_with_x_not_found(ortho, 7)

print(f'Species with one missing ortholog {cols_with_one_not_found}')
print(f'Species with two missing ortholog {cols_with_two_not_found}')
print(f'Species with three missing ortholog {cols_with_three_not_found}')
print(f'Species with four missing ortholog {cols_with_four_not_found}')
print(f'Species with five missing ortholog {cols_with_five_not_found}')
print(f'Species with six missing ortholog {cols_with_six_not_found}')
print(f'Species with seven missing ortholog {cols_with_seven_not_found}')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 600

# Compute count of "Not found" entries for each column
not_found_counts = ortho.eq('Not found').sum()

# Count columns with x "Not Found"
counts = [sum(not_found_counts == i) for i in range(1, 8)]

# Adding counts as text above each bar
for i, count in enumerate(counts):
    plt.text(i + 1, count + 0.5, str(count), ha='center', va='bottom')

# Plotting
plt.bar(range(1, 8), counts, color='orange', alpha=0.7)
plt.xlabel('Number of missing orthologs')
plt.ylabel('Number of species')
plt.title('Histogram of missing orthologs per species')
plt.xticks(range(1, 8))
plt.yticks(range(0, 700, 100))
plt.grid(alpha=0.1, linestyle='--')
plt.show()
```

```python
d = {
    'NP_009673.1': [],  # LYS2
    'NP_010290.3': [],  # TRP1
    'NP_010851.1': [],  # CAN1
    'NP_010893.3': [],  # URA3
    'NP_012965.3': [],  # GAP1
    'NP_013406.1': [],  # MET17
    'NP_015387.1': []   # FCY1
}

names = []

for tsv in [e + '.tsv' for e in ortho.columns[1:].tolist()]:
    names.append(tsv)
    df = pd.read_csv(f'outputs/{tsv}', sep='\t', names=['scer_gene', 'ortho', 'identity', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'])

    for gene, l in d.items():
        if gene in df['scer_gene'].values:
            d[gene].append(df[df['scer_gene'] == gene]['identity'].values[0])
        else:
            d[gene].append(-1)
            
total_missing = 0
for gene in d.keys():
    length = len([e for e in d[gene] if e != -1])
    print(f'Gene {gene} {acc_to_name[gene]} is missing its ortholog in {num_files - length} species')
    total_missing += num_files - length

print(f'Total missing orthologs {total_missing}')
```

# Sensitivity analysis

Find out the ranges of ortholog identity for best hits in each species

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
import numpy as np

keys = list(d.keys())
values = list(d.values())

# Setting up the subplots
fig, axs = plt.subplots(nrows=len(keys), figsize=(10, 30))

# Determining the global min and max values for bin range
global_min = min([min(v) for v in values])
global_max = max([max(v) for v in values])

# Looping through each key and generating histograms with fixed bins and displaying counts above each bar
for i, (key, value) in enumerate(d.items()):
    # Plot histogram
    counts, bins, patches = axs[i].hist(value, bins=np.arange(0, 100, 10), alpha=0.7, color='orange')
    
    # Add counts above each bar
    for count, bin, patch in zip(counts, bins, patches):
        height = patch.get_height()
        axs[i].text(bin + 5, height + 0.2, str(int(count)), ha='center', va='bottom', fontsize=9, color='black')
    
    axs[i].set_title(f"{acc_to_name[key]} Identity of Best Ortholog per Species")
    axs[i].set_ylabel("Species Frequency")
    axs[i].set_xlabel("Ortholog Identity Percentage")
    axs[i].set_xlim(global_min, global_max)
    axs[i].set_ylim(0, 2500)
    axs[i].set_xticks(np.arange(0, 101, 10))
    axs[i].grid(alpha=0.1, linestyle='--')

# Adjusting layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
for gene in d.keys():
    low_quality = len([e for e in d[gene] if e >= 0 and e < 30])
    length = len([e for e in d[gene] if e != -1])

    print(f'Gene {gene} has {low_quality}/{length} low quality (< 30% identity) orthologs')
```

## Mark orthologs with low (< 30%) identity as "Not found"

```python
ortho_high_identity = ortho.copy(deep=True)

for species in ortho.columns[1:].tolist():
    df = pd.read_csv(f'outputs/{species}.tsv', sep='\t', names=['scer_gene', 'ortho', 'identity', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'])
    low_identities = df[df['identity'] < 30]['scer_gene'].tolist()
    ortho_high_identity.loc[ortho_high_identity['Orthogroup'].isin(low_identities), species] = 'Not found'

ortho_high_identity
```

```python
# Compute count of "Not found" entries for each column
not_found_counts = ortho_high_identity.eq('Not found').sum()

# Count columns with x "Not Found"
counts = [sum(not_found_counts == i) for i in range(1, 8)]

# Adding counts as text above each bar
for i, count in enumerate(counts):
    plt.text(i + 1, count + 0.5, str(count), ha='center', va='bottom')

# Plotting
plt.bar(range(1, 8), counts, color='orange', alpha=0.7)
plt.xlabel('Number of missing orthologs')
plt.ylabel('Number of species')
plt.title('Histogram of missing orthologs per species')
plt.xticks(range(1, 8))
plt.yticks(range(0, 801, 100))
plt.grid(alpha=0.1, linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

source_counts = input_species['source'].value_counts()

cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]

# Function to display both raw counts and percentages
def func(pct, allvals):
    absolute = int(round(pct / 100.0 * sum(allvals)))
    return "{:.1f}% ({:d})".format(pct, absolute)

# Plotting the pie chart with shadow, white text color, and both raw counts and percentages
plt.figure(figsize=(10, 6))
source_counts.plot(kind='pie', autopct=lambda p: func(p, source_counts), colors=colors)
plt.title('Distribution of Finalized Sources')
plt.ylabel('')  # Removing the default y-label
plt.show()
```
