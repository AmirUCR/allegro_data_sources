## A7 CDS 1MM

```python
df_a7_1mm_cds = pd.read_csv('output/km_a7_1mm_cds_alignment.sam', sep='\t',
                names=['query_name', 'strand', 'reference_name', 'mapping_position',
                       'Sequence', 'mapping_quality', 'idk', 'mismatch'])
df_a7_1mm_cds
```

# rows with mismatches occuring after the 10th base should be dropped

```python
# Create a new column called mismatch_position and extract the mismatch position digits from
# the mismatch column. Then fill NaN values with 0 because NaN means no mismatch.
# Then drop rows where the mismatch occurs after the 10th base.
df_a7_1mm_cds['mismatch_position'] = df_a7_1mm_cds['mismatch'].str.extract(r'(\d+):.*')[0]
df_a7_1mm_cds['mismatch_position'].fillna(0, inplace=True)
df_a7_1mm_cds['mismatch_position'] = df_a7_1mm_cds['mismatch_position'].astype(int)
df_a7_1mm_cds[df_a7_1mm_cds['mismatch_position'] < 10]
```
