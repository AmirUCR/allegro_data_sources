```python
import pandas as pd
import matplotlib.pyplot as plt


species = ['yarrowia_lipolytica', 'komagataella_phaffii',
           'kluyveromyces_marxianus', 'saccharomyces_cerevisiae']

def get_total_guides(species_name, track):
    df = pd.read_csv(f'{species_name}/output/{track}/{track}_lib_offtargets.csv')

    return len(df)

def get_offtarget_counts(species_name, track):
    df = pd.read_csv(f'{species_name}/output/{track}/{track}_lib_offtargets.csv')

    bin_0mm_cds = df.astype(bool).sum(axis=0)['0mm_cds_offtargets']
    bin_1mm_cds = df.astype(bool).sum(axis=0)['1mm_cds_offtargets']
    bin_2mm_cds = df.astype(bool).sum(axis=0)['2mm_cds_offtargets']
    bin_0mm_genome = df.astype(bool).sum(axis=0)['0mm_genome_offtargets']
    bin_1mm_genome = df.astype(bool).sum(axis=0)['1mm_genome_offtargets']
    bin_2mm_genome = df.astype(bool).sum(axis=0)['2mm_genome_offtargets']

    return bin_0mm_cds, bin_1mm_cds, bin_2mm_cds, bin_0mm_genome, bin_1mm_genome, bin_2mm_genome
```

```python
import numpy as np

plt.rcParams['figure.dpi'] = 300

marxianus_total_guides = get_total_guides('kluyveromyces_marxianus', 'a7')
phaffii_total_guides = get_total_guides('komagataella_phaffii', 'a7')
lipolytica_total_guides = get_total_guides('yarrowia_lipolytica', 'a7')
cerevisiae_total_guides = get_total_guides('saccharomyces_cerevisiae', 'a7')

labels = ['0', '1', '2']
values1_cds = get_offtarget_counts('kluyveromyces_marxianus', 'a7')[:3]
values2_cds = get_offtarget_counts('komagataella_phaffii', 'a7')[:3]
values3_cds = get_offtarget_counts('yarrowia_lipolytica', 'a7')[:3]
values4_cds = get_offtarget_counts('saccharomyces_cerevisiae', 'a7')[:3]

values1_genome = get_offtarget_counts('kluyveromyces_marxianus', 'a7')[3:]
values2_genome = get_offtarget_counts('komagataella_phaffii', 'a7')[3:]
values3_genome = get_offtarget_counts('yarrowia_lipolytica', 'a7')[3:]
values4_genome = get_offtarget_counts('saccharomyces_cerevisiae', 'a7')[3:]

# Position of bars on x-axis and bar width
ind = np.arange(len(labels))
width = 0.2  # adjust width to fit four bars

def add_labels(bars, ax, color):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color=color)

def create_bar_plot(ax, values1, values2, values3, values4, title):
    km_bars = ax.bar(ind - 3*width/2, values1, width, label=f'K. marxianus ({marxianus_total_guides} Total A7 Guides)', color='#ffd166')
    kphaffii_bars = ax.bar(ind - width/2, values2, width, label=f'K. phaffii ({phaffii_total_guides} Total A7 Guides)', color='#ff2647')
    ylipolytica_bars = ax.bar(ind + width/2, values3, width, label=f'Y. lipolytica ({lipolytica_total_guides} Total A7 Guides)', color='#06d6a0')
    scerevisiae_bars = ax.bar(ind + 3*width/2, values4, width, label=f'S. cerevisiae ({cerevisiae_total_guides} Total A7 Guides)', color='#3a86ff')

    # Add counts above each bar
    add_labels(km_bars, ax, color='#ffd166')
    add_labels(kphaffii_bars, ax, color='#ff2647')
    add_labels(ylipolytica_bars, ax, color='#06d6a0')
    add_labels(scerevisiae_bars, ax, color='#3a86ff')

    ax.set_ylabel('Number of Guides')
    ax.set_xlabel('Exact Number of Mismatches')
    ax.set_title(f'[A7] Number of Guides Having Any Number of Off-Targets in the {title}')
    ax.set_xticks(ind)
    ax.set_yticks([n for n in range(5)])
    ax.set_xticklabels(labels)
    ax.grid(alpha=0.1)
    ax.legend(loc='upper left')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))  # 1 row, 2 columns, larger figure size

# Create the first plot
create_bar_plot(ax1, values1_cds, values2_cds, values3_cds, values4_cds, 'CDS')

# Create the second plot, identical to the first
create_bar_plot(ax2, values1_genome, values2_genome, values3_genome, values4_genome, 'Genome')

# Show the plots
plt.tight_layout()  # This ensures that the plots do not overlap
plt.show()
```
