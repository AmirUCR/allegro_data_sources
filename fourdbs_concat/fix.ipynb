```python
import pandas as pd

df = pd.read_csv('fourdbs_input_species.csv')

df
```

```python
# Replace non-unicode with underscores. E.g., punctiations
df.species_name = df.species_name.str.replace(r'[^\w\s]+', '_', regex=True)
df.original_name = df.original_name.str.replace(r'[^\w\s]+', '_', regex=True)
df.cds_file_name = df.cds_file_name.str.replace(r'[^\w\s]+', '_', regex=True)
df.genome_file_name = df.genome_file_name.str.replace(r'[^\w\s]+', '_', regex=True)

# Replace consecutive underscores with one
df.species_name = df.species_name.str.replace(r'(_)\1+', '_', regex=True)
df.original_name = df.original_name.str.replace(r'(_)\1+', '_', regex=True)
df.cds_file_name = df.cds_file_name.str.replace(r'(_)\1+', '_', regex=True)
df.genome_file_name = df.genome_file_name.str.replace(r'(_)\1+', '_', regex=True)

# Remove trailing _
df.original_name = df.original_name.str.replace(r'_$', '', regex=True)

# Fix extensions
df.cds_file_name = df.cds_file_name.str.replace(r'_fna', '.fna', regex=True)
df.genome_file_name = df.genome_file_name.str.replace(r'_fna', '.fna', regex=True)
```

```python
species_with_six_missing_ortho = ['anncaliia_algerae',
'dictyocoela_muelleri',
'dictyocoela_roeselum',
'edhazardia_aedis',
'enteropsectra_breve',
'mitosporidium_daphniae_ugp3',
'nematocida_displodere',
'nematocida_homosporus',
'ordospora_colligata',
'pancytospora_philotis',
'pseudoloma_neurophilia_str_mk1',
'tubulinosema_ratisbonensis',
'vavraia_culicis',
'vittaforma_corneae']

species_with_seven_missing_ortho = ['cucumispora_dikerogammari',
'enterocytozoon_bieneusi_h348',
'enterocytozoon_hepatopenaei_str_th1',
'enterospora_canceri_str_gb1',
'hamiltosporidium_magnivora',
'hamiltosporidium_tvaerminnensis',
'hepatospora_eriocheir',
'nematocida_ausubeli',
'nematocida_major',
'nematocida_minor',
'nematocida_parisii',
'nematocida_sp_awrm78',
'nosema_apis_brl_01',
'pancytospora_epiphaga',
'spraguea_lophii_42_110',
'talaromyces_borbonicus_cbs_141340',
'trachipleistophora_hominis',
'vairimorpha_apis']

rows_to_drop = list()
rows_to_drop.extend(species_with_six_missing_ortho)
rows_to_drop.extend(species_with_seven_missing_ortho)

df = df[~df['original_name'].isin(rows_to_drop)]
df = df.reset_index(drop=True)
df
```

```python
df.to_csv('fourdbs_input_species.csv', index=False)
```

```python
import os
import re
import shutil

if not os.path.exists('removed'):
    os.mkdir('removed')

for file in os.listdir('cds'):
    new_name = re.sub(r'[^\w\s]+', '_', file)  # Replace non unicode chars with _
    new_name = re.sub(r'(_)\1+', '_', new_name)  # Replace consecutive _
    new_name = re.sub(r'_fna', '.fna', new_name)  # Fix extensions
    old_path = os.path.join('cds', file)
    new_path = os.path.join('cds', new_name)
    os.rename(old_path, new_path)
    
    if new_name[0:new_name.index('_cds.fna')] in rows_to_drop:
        shutil.move(new_path, f'removed/{new_name}')

for file in os.listdir('genomes'):
    new_name = re.sub(r'[^\w\s]+', '_', file)
    new_name = re.sub(r'(_)\1+', '_', new_name)
    new_name = re.sub(r'_fna', '.fna', new_name)
    old_path = os.path.join('genomes', file)
    new_path = os.path.join('genomes', new_name)
    os.rename(old_path, new_path)

    if new_name[0:new_name.index('_genomic.fna')] in rows_to_drop:
        shutil.move(new_path, f'removed/{new_name}')

for file in os.listdir('proteomes'):
    new_name = re.sub(r'[^\w\s]+', '_', file)
    new_name = re.sub(r'(_)\1+', '_', new_name)
    new_name = re.sub(r'_faa', '.faa', new_name)
    old_path = os.path.join('proteomes', file)
    new_path = os.path.join('proteomes', new_name)
    os.rename(old_path, new_path)

    if new_name[0:new_name.index('.faa')] in rows_to_drop:
        shutil.move(new_path, f'removed/{new_name}')
```
