```python
import pandas as pd

ortho = pd.read_csv("Orthogroups.tsv", sep='\t')
ortho
```

```python
ortho['yarrowia_lipolytica']
```

```python
# Find IDs corresponding to "Not Found" for each column
not_found_ids = {}
for col in ortho.columns:
    if col != 'Orthogroup':  # Exclude the 'Orthogroup' column from our search
        ids = ortho['Orthogroup'][ortho[col] == 'Not found'].tolist()
        if ids:
            not_found_ids[col] = ids

print(not_found_ids)
```

```python
def find_cols_with_x_not_found(df, x):
    return df.columns[df.eq('Not found').sum() == x].tolist()
```

```python
# Identify columns where only one row is "Not Found"
cols_with_one_not_found = find_cols_with_x_not_found(ortho, 1)
cols_with_two_not_found = find_cols_with_x_not_found(ortho, 2)
cols_with_three_not_found = find_cols_with_x_not_found(ortho, 3)
cols_with_four_not_found = find_cols_with_x_not_found(ortho, 4)
cols_with_five_not_found = find_cols_with_x_not_found(ortho, 5)
cols_with_six_not_found = find_cols_with_x_not_found(ortho, 6)
cols_with_seven_not_found = find_cols_with_x_not_found(ortho, 7)

print(f'Species with one missing ortholog {cols_with_one_not_found}')
print(f'Species with two missing ortholog {cols_with_two_not_found}')
print(f'Species with three missing ortholog {cols_with_three_not_found}')
print(f'Species with four missing ortholog {cols_with_four_not_found}')
print(f'Species with five missing ortholog {cols_with_five_not_found}')
print(f'Species with six missing ortholog {cols_with_six_not_found}')
print(f'Species with seven missing ortholog {cols_with_seven_not_found}')

cols_with_one_not_found.insert(0, 'Orthogroup')
```

```python
ortho[cols_with_one_not_found]
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 600

# Compute count of "Not found" entries for each column
not_found_counts = ortho.eq('Not found').sum()

# Count columns with x "Not Found"
counts = [sum(not_found_counts == i) for i in range(1, 8)]

# Plotting
plt.bar(range(1, 8), counts, color='orange')
plt.xlabel('Number of missing orthologs')
plt.ylabel('Number of species')
plt.title('Histogram of missing orthologs per species')
plt.xticks(range(1, 8))
plt.show()
```
